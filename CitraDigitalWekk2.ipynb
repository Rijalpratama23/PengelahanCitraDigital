```python
import cv2
from google.colab.patches import cv2_imshow

# membaca citra/gambar dari file
gambar = cv2.imread('poto1.jpg')
# menampilkan citra/gambar
cv2_imshow(gambar)
cv2.waitKey(0)
```

```python
# check ukuran pixel
if gambar is not None:
  tinggi, lebar, channel = gambar.shape
  print(f"Tinggi: {tinggi}, Lebar: {lebar}, Channel: {channel}")
```

```python
# mengubah ke GrayScale
gray_gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_gambar)
cv2.waitKey(0)
cv2.destroyAllWindows
```

Di bawah ini merupakan Tugas;

```python
color1_gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2HSV)
cv2_imshow(color1_gambar)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
color2_gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2LAB)
cv2_imshow(color2_gambar)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
color3_gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2LUV)
cv2_imshow(color3_gambar)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
# Resize Pixel
resized_gambar = cv2.resize(gambar, (400, 400))
cv2_imshow(resized_gambar)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
cv2.imwrite('gambar_resize.jpg', resized_gambar)
```

```python
cv2.imwrite('gambar_gray.jpg', gray_gambar)
```
